**📈 Heatmap (Correlation Matrix)**

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Generate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

The error you're encountering is due to the presence of non-numeric columns in the DataFrame when trying to compute the correlation matrix using df.corr(). The correlation matrix can only be computed for numeric data types.

To resolve this issue, you can either: **bold text**

1. **Exclude non-numeric columns** from the DataFrame before calculating the correlation matrix.

2. **Convert categorical variables to numeric** using encoding techniques (like one-hot encoding) if you want to include them in the correlation analysis.

# **Solution: Exclude Non-Numeric Columns**
Here’s how you can modify the code to exclude non-numeric columns before generating the correlation heatmap:

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number]) #This line selects only the numeric columns from the DataFrame.
print(numeric_df)
# Generate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**6. Advanced Visualizations (Seaborn & Plotly)**

**📊 Pair Plot (All Numerical Relationships)**

```python
# Select meaningful columns for correlation analysis
meaningful_columns = ['age', 'grade_level',
                      'attendance_rate', 'study_hours', 'gender','final_result']
df_meaningful = df[meaningful_columns].copy()

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df_meaningful, drop_first=True)

# Generate the correlation matrix
correlation_matrix = df_encoded.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap with Meaningful Columns')
plt.show()
```

**📊 Interactive Plot (Plotly)**

```python
fig = px.scatter_3d(df, x='age', y='gender', z='grade_level', color='math_score')
fig.show()
```
